# Matplotlib and Seaborn

```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading titanic dataset
df = sns.load_dataset('titanic')
df
```

```python
# Checking all the datatypes and non null values
df.info()
```

## Scatter Plot

```python
# Making a scatter plot of age vs fare by class
sns.scatterplot(df, x='age', y='fare', hue='class', palette='Set2')

# Giving title to the chart
plt.title('Age vs Fare by Class')

# Labels
plt.xlabel('Age')
plt.ylabel('Fare')

# show plot
plt.show()
```

```python
# Setting the size of the figure
plt.figure(figsize = (13, 5))

# Making a scatter plot of age vs fare by survival
sns.scatterplot(df, x='age', y='fare', hue='survived', palette='Set2')

# Giving title to the chart
plt.title('Age vs Fare by Survival')

# Labels
plt.xlabel('Age')
plt.ylabel('Fare')

# show plot
plt.show()
```

```python
# Selecting the size on our own
plt.figure(figsize=(10, 6))

# Making the scatterplot for age vs fare by survival
# alpha is the transparency of colour
# sizes=(20, 200) sets the minimum and maximum size of the dots (points) in the scatter plot
sns.scatterplot(df,
    x='age',
    y='fare',
    hue='survived',
    palette='viridis',
    sizes=(50, 300),
    alpha=0.5
)

# Giving title to the chart
plt.title('Age vs Fare by Survival')

# show plot
plt.show()
```

```python
# Selecting the size on our own
plt.figure(figsize=(10, 6))

# Making scatterplot
sns.scatterplot(df,
    x='age',
    y='fare',
    hue='survived',
    palette='dark',
    sizes=(50, 300),
    alpha=0.5
)

# Giving title to the chart
plt.title('Age vs Fare by Survival(Dark Color)')

# show plot
plt.show()
```

```python
# Selecting the size on our own
plt.style.use('dark_background')
plt.figure(figsize=(10, 6))

# Making scatterplot
sns.scatterplot(df,
    x='age',
    y='fare',
    hue='survived',
    palette='deep',
    sizes=(50, 300),
    alpha=0.5
)

# Giving title to the chart
plt.title('Age vs Fare by Survival(Black Background)')

# show plot
plt.show()
```
