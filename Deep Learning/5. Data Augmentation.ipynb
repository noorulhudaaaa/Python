# Data Augmentation
Data Augmentation is like giving your model more practice by creating new training images from old ones, using small smart changes.

## Why is this helpful?

- Makes your model see more examples
- Helps your model generalize better (do well on new images)
- Reduces overfitting (model memorizing instead of learning)

```python
# Importing necessary Libraries
import tensorflow
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array, array_to_img
```

```python
# rotation_range=30	Turn the image left/right a little (up to 40 degrees)
# width_shift_range=0.3	Move the image left/right by 20% of the width
# height_shift_range=0.3	Move the image up/down by 20% of the height
# shear_range=0.3	Slant the image (like stretching sideways)
# zoom_range=0.3	Zoom in or out a little
# horizontal_flip=True	Flip the image left-to-right (like a mirror)
# fill_mode='nearest'	If any empty space appears after changes, fill it with the nearest

gen = ImageDataGenerator(
    rotation_range = 30,
    width_shift_range = 0.3,
    height_shift_range = 0.3,
    shear_range = 0.3,
    zoom_range = 0.3,
    horizontal_flip = True,
    fill_mode = 'nearest'
)
```

```python
# Loading our image
image = load_img('/content/cute-dog-jpg_1072857-373.jpg')
image

# This is PIL Image not numpy array
```

```python
# Converting the image to numpy array
x = img_to_array(image)
x.shape
```

```python
# Now we need an imae that has height, width, channels as well as batch size
# We will perform concatenation by adding 1 in the front (for batch size)
x = x.reshape((1, ) + x.shape)
x.shape

# Now we have a batch size along everything else
```

```python
# Setting the index of i at 0 first for loop
i = 0

# flow method is used to generate different random images
for batch in gen.flow(x, batch_size = 1,
                 save_to_dir = '/content/sample_data/Dog Images',
                 save_prefix = 'Dog',
                 save_format = 'jpeg'):
  i += 1
  if i >= 20:
     break
```
