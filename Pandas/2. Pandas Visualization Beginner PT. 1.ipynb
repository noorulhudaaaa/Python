# Pandas Visualization Pt. 1

```python
# Importing all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the file
df = pd.read_csv('Ice Cream Ratings.csv')

# Setting date column as index
df = df.set_index('Date')

# or df.set_index('Date', inplace = True)

df
```

## Line Plot

```python
# Here it will make a line chart by default
df.plot()
```

```python
# We can make our charts visually better
print(plt.style.available)

# Now can use any of them
plt.style.use('dark_background')
```

```python
# Even if we specify line chart, it will be same
# df.plot(kind = 'line')

df.plot.line()

# They both give the same result aswell
```

```python
# We can make a line graph for a specific column aswell
df['Flavor Rating'].plot.line( title = 'Ice Creamm Rating', ylabel = 'Score')
```

```python
# Creating line graphs for three of these columns separately in one visualization
df.plot(kind = 'line', subplots = True)
```

## Area Plot

```python
# Creating an area plot
df.plot.area(title = 'Area Plot', figsize = (10, 5))
```

## Bar Plot

```python
# Making a stacked bar chart
df.plot.bar(stacked = True)
```

```python
# Another way to make it
# Similary can add stacked = True after kind to make it stacked
df.plot(kind = 'bar')

# Can make it visually better by adding things
```

```python
# Making a bar chart of a specific column
df['Texture Rating'].plot(kind = 'bar', title = 'Bar Chart', ylabel = 'Rating')
```

```python
# Making a horizontal stacked bar chart
df.plot.barh(stacked = True)
```

## Scatter Plot

```python
# Making a scatter plot of Flavor Rating Vs Texture Rating
df.plot.scatter(x = 'Flavor Rating', y = 'Texture Rating', s = 100, c = 'green', title = 'Scatter Plot')

# s is te size of the points and c is the colour of the points
```

## Histogram

```python
df['Texture Rating'].plot.hist(bins = 15)
```

```python
# Bins are 10 by default
# alpha decides the transparenscy of the colour
df.plot.hist(bins = 20, alpha = 0.9)
```

## Box Plot

```python
# Creating a box plot
df.plot.box(color = 'purple', title = 'Box Plot')
```

## Pie Chart

```python
# Creating a pie chart
df.plot.pie(y = 'Flavor Rating', title = 'Pie Chart', figsize = (10, 7))
```
