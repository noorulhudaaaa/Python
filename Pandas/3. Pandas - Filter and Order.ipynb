# Filtering and Ordering in Pandas

```
# Importing library for panda
import pandas as pd
```

```
# Loading csv file using pandas
df = pd.read_csv('world_population.csv')
df
```

## Filtering By Columns

```
# Filtering
# We are displaying only those rows where rank is below or equal to 5
df[df['Rank'] <= 5]
```

```
# Filtering
# We are displaying only the rows where continent is North America
df[df['Continent'] == 'North America']
```

```
# If you want to add two conditions
# Displaying rows where country is china and india
df[(df['Country'] == 'China') | (df['Country'] == 'India')]
```

```
# Both these give same output
specific = ['China', 'India']
df[df['Country'].isin(specific)]
```

```
# This also gives the same output as other two
df[df['Country'].str.contains('China|India', na=False)]
```

## Filtering By Index

```
# Now we are makiing one of our column (country) as index
df1 = df.set_index('Country')
df1
```

```
# Getting the only required columns
df1.filter(items = ['Continent', 'CCA3', 'Rank'], axis = 1)
```

```
# DSo this wa we can filter out all the information specifically for yemen country
df1.filter(items = ['Yemen'], axis = 0)
```

```
# Can also do this if we want it for more countries
df1.filter(like = 'United', axis = 0)
```

#### loc and iloc

```
# Since we have made country as index, we ccan now get all columns data of any country using loc
df1.loc['Yemen']
```

```
# Getting all the columns data for United States
df1.loc['United States']
```

```
# This will check and give all the columns data for index 3 excluding the country
df1.iloc[3]
```

## Ordering

```
df
```

```
# All the rows where rank is less than equal to 10
# Orderinng it by rank, in ascending order
df[df['Rank'] <= 10].sort_values(by='Rank')
```

```
# All the rows where rank is less than equal to 10
# Orderinng it by rank, in ascending order
df[df['Rank'] <= 10].sort_values(by='Rank', ascending = False)
```

```
# You can add 2 thing or more in order by
# However it follows the order of the first provided column and ignores the second one
# asecending = true gives ascending order and false gives descending order
df[df['Rank'] <= 10].sort_values(by=['Continent', 'Rank'], ascending = False)
```
