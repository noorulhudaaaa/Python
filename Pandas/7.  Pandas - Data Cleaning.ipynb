# Pandas - Data Cleaning

```python
# Importing libraries
import pandas as pd
```

```python
# Reading excel file
df = pd.read_excel('Customer Call List.xlsx')
df
```

## Dropping Duplicates

```python
# First of all, we will remove all the duplicates
df = df.drop_duplicates()
df
```

## Dropping Unnecessary Columns

```python
# Now we will drop the unnecessary column if any i.e Non Uselful column
df = df.drop(columns = 'Not_Useful_Column')
df

# If multiple columns are to be deleted
# df.drop(columns = ['Not_Useful_Column', ' Do_Not_Contact'])
```

## Data Standardization

```python
# now we will standardize our data, if any characters of extra saces present in any rows, we will remove
# df['Last_Name'] = df['Last_Name'].str.lstrip('/')
# df['Last_Name'] = df['Last_Name'].str.lstrip('...')
# df['Last_Name'] = df['Last_Name'].str.rstrip('...')
# df['Last_Name'] = df['Last_Name'].str.rstrip('_')

# An easier way out
df['Last_Name'] = df['Last_Name'].str.strip('/_.')
df
```

```python
# Now the Phone_Number column numbers in differentb formats
# First we will remove any character present in the number
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
df
```

```python
# Now we will add - right after each 3 digits and then last 4 digits together
# needs to be converted in string first becuase we have string values in columns aswell like na
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))

df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
#  Now replace nan and na with empty space
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
df
```

## Splitting a column into more

```python
# Splitting adress column into 3
df[["Street_Adress", "State", "Zip_Code"]] = df['Address'].str.split(',', n=2, expand = True)
df
```

```python
# Now for the paying customers there are yes/no and y/no, so have to make them same
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df
```

```python
# Doing the same for Do_Not_Contact
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df
```

## Handling Empty Values

```python
# Now putting empty space inplace of N/a
df = df.replace('N/a', '')
df
```

```python
# However the same doesnt work for NaN because its string
# We use another way
df = df.fillna('')
df

# This will remove all the None and NaN values  and make them empty
```

## Dropping rows according to Requirement

```python
# No we want to drop those rows out where people said Yes for donnot contact
for x in df.index:
  if df.loc[x, 'Do_Not_Contact'] == 'Y':
    df.drop(x, inplace = True)

df
```

```python
# No we want to drop those rows out where people have no contact number
for x in df.index:
  if df.loc[x, 'Phone_Number'] == '':
    df.drop(x, inplace = True)

df

#  df.dropna(subset = "Phone_Number"), inplace - True would also do same
```
