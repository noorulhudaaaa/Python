# Pandas - Exploratory Data Analysis

```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading our csv file
df = pd.read_csv('world_population.csv')
df
```

```python
# We converting values to float
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df
```

```python
# Checking Information
df.info()
```

```python
# High level view of columns
df.describe()
```

```python
# Calculating total null values in each column
df.isnull().sum()
```

```python
# Checking unique values in each column
df.nunique()
```

```python
# Sorting values for 2022 population in ascending order
# Printing the 1st 10 rows
df.sort_values(by='2022 Population').head(10)
```

```python
 # Sorting values for 2022 population in descending order
# Printing the 1st 10 rows
df.sort_values(by='2022 Population', ascending = False).head(10)

# This gives top 10 most populated capitals in 2022
```

```python
 # Sorting values for 2022 population in descending order
# Printing the 1st 10 rows
df.sort_values(by='World Population Percentage', ascending = False).head(10)

# This gives top 10 countrieswith the most population percentage
```

```python
# Checking corr() of the Numerical Values
df.corr(numeric_only=True)
```

```python
# Visualizing this correlation using heatmap
# Annot = true, gives numbers inside boxes
# Selecting the figure size
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(numeric_only=True), annot=True)

# showing heatmap
plt.show()
```

```python
# Grouping by the Continent and applying mean() aggregate function
df.groupby('Continent').mean(numeric_only = True)
```

```python
# If want to look for a specific category data
df[df['Continent'].str.contains('Oceania')]
```

```python
# Grouping by the Continent and applying mean() aggregate function
# sorting by the 2022 polpulation in descending order
# Which Continent has grown the fastest in 2022
df2 = df.groupby('Continent').mean(numeric_only = True).sort_values(by='2022 Population', ascending = False)
df2
```

```python
# Visualizing this
df2.plot()
```

```python
# now all other columns will be on x axis
df3 = df2.transpose()
df3
```

```python
# Now making plot with all columns on x axis
df3.plot(figsize=(15, 7))

# Showing plot
plt.show()

# We just want the population columns
```

```python
df4 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by='2022 Population', ascending = False)
df4
```

```python
df5 = df4.transpose()
df5
```

```python
# Now making plot with all mentioned columns on x axis
df5.plot(figsize=(15, 7))

# Showing plot
plt.show()

# We just want the population columns
```

```python
# Making a boxplot
df.boxplot(figsize=(20,10))
```

```python
# Now making plot with all columns on x axis
df4.boxplot(figsize=(15, 7))

# Showing plot
plt.show()

# We just want the population columns
```
