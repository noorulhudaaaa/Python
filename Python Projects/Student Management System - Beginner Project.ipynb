# Student Management System Project

A simple command-line project to manage student records. Great for beginners!

---


## Features and Concepts Used


| **Features**                              | **Python Concepts Used**         |
|------------------------------------------|----------------------------------|
| Add student (auto percentage & grade)    | Input/Output functions           |
| View all student records                 | Lists and Dictionaries           |
| Search by name or roll number            | Loops and Conditionals           |
| Sort students by marks (high to low)     | Sorting with lambda functions    |
| Remove a student by roll number          | Functions   |

---



```python
students = []
 # defining function for grade
def gradeStu(percentage):
  if percentage >= 90:
    return "A+"
  elif percentage >= 85 and percentage <= 89:
    return "A"
  elif percentage >= 80 and percentage <= 84:
    return "B+"
  elif percentage >= 75 and percentage <= 79:
    return "B"
  elif percentage >= 70 and percentage <= 74:
    return "C+"
  elif percentage >= 65 and percentage <= 69:
    return "C"
  elif percentage >= 60 and percentage <= 64:
    return "D+"
  elif percentage >= 50 and percentage <= 59:
    return "D"
  else:
    return "F"
```

```python
# Adding student data
# Defining function for adding students data

def add_student():

    # Getting input from user
    name = input("Enter student name: ")
    roll_no = input("Enter roll number: ")
    marks = float(input("Enter marks: "))
    totalMarks = float(input("Enter total marks: "))

    percentage = int((marks / totalMarks) * 100)
    grade = gradeStu(percentage)                      # calculating grade

    #  Creating dictionary for student details
    student = {"name": name, "roll_no": roll_no, "marks": (marks), "totalMarks": (totalMarks), "percentage": (percentage), "grade": grade}

    #  Adding new student to list
    students.append(student)
    print("Student added successfully!\n")
```

```python
# view students's data

def view_students():

    # Checking if students list is empty
    if not students:
        print("No student records found.\n")
        return

    # Diplay each student's details
    for idx, student in enumerate(students, start=1):
        print(f"{idx}. Name: {student['name']}, Roll No: {student['roll_no']}, "
        f"Marks: {student['marks']}, Total Marks: {student['totalMarks']}, "
              f"Percentage: {student['percentage']}%, Grade: {student['grade']}")
    print()
```

```python
# Searh student using name or roll number

def search_student():
    keyword = input("Enter name or roll number to search: ")
    found = False

    # Checking the students list
    for student in students:
        if student["name"] == keyword or student["roll_no"] == keyword:

          # Display if found
            print(f"Found: Name: {student['name']}, Roll No: {student['roll_no']}, "
                  f"Marks: {student['marks']}, Total Marks: {student['totalMarks']}, "
                  f"Percentage: {student['percentage']}%, Grade: {student['grade']}")
            found = True   # updating flag to true if found

    if not found:
        print("Student not found.\n")
```

```python
# Sorted students on basis of marks

# Checking if list is empty
def sortingStu():
    if not students:
        print("No student records found.\n")
        return

    # Sorting list by marks of each student from top to bottom
    sorted_list = sorted(students, key=lambda x: x["marks"], reverse=True)

    print("\nStudents sorted by marks (highest to lowest):\n")
    for idx, student in enumerate(sorted_list, start=1):           # Displaying the sorted list
        print(f"{idx}. Name: {student['name']}, Roll No: {student['roll_no']}, "
              f"Marks: {student['marks']}, Total Marks: {student['totalMarks']}, "
              f"Percentage: {student['percentage']}%, Grade: {student['grade']}")
    print()

```

```python
# Remove student's data

def remove_student():
  # asking user to enter roll no
    roll_no = input("Enter roll number of student to remove: ")
    for student in students:
        if student["roll_no"] == roll_no:

          # if match found, student data will be removed
            students.remove(student)
            print("Student removed successfully.\n")
            return


      # if no match founf, it notifies the user
    print("Student not found.\n")
```

```python
def main():
    while True:
        print("------ Student Management System ------")
        print("1. Add Student")
        print("2. View All Students")
        print("3. Search Student")
        print("4. Sort Students by Marks")
        print("5. Remove Student")
        print("6. Exit")


        # Getting users choice
        choice = input("Enter your choice: ")

        if choice == '1':
            add_student()
        elif choice == '2':
            view_students()
        elif choice == '3':
            search_student()
        elif choice == '4':
            sortingStu()
        elif choice == '5':
            remove_student()
        elif choice == '6':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please try again.\n")

# Run the program
main()
```
